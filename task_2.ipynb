Create a basic bar chart or line chart using a popular Python visualization library like Matplotlib or Seaborn. Visualize key insights from the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Iris.csv')
```

```python
plt.bar(data["Species"], data["SepalLengthCm"])
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.title("Sepal Length Comparison by Species")
plt.show()
```

```python
data.hist(edgecolor='black', linewidth=1.2)
fig=plt.gcf()
fig.set_size_inches(12,12)
plt.show()
```
